```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def threshold_image(input_path, output_path, threshold_values):
    try:
        # Read the input image
        img = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
        original = cv2.imread(input_path, cv2.IMREAD_COLOR)
        img_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)


        # Apply thresholding with three different values
        thresholded_images = [cv2.threshold(img, threshold, 255, cv2.THRESH_BINARY)[1] for threshold in threshold_values]

        # Plotting
        fig, axs = plt.subplots(2, 2, figsize=(10, 8))

        # Display original image
        axs[0, 0].imshow(img_rgb, cmap='gray')
        axs[0, 0].set_title('Original Image')

        # Display thresholded images
        for i, (thresholded_img, threshold_value) in enumerate(zip(thresholded_images, threshold_values), start=1):
            axs[i // 2, i % 2].imshow(thresholded_img, cmap='gray')
            axs[i // 2, i % 2].set_title(f'Threshold {i}: {threshold_value}')
            cv2.imwrite(output_path.format(i), thresholded_img)

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    input_path = input('Enter input file name: ')
    output_path = input('Enter output file name: ')
    threshold_values = [100, 150, 200]

    threshold_image(input_path, output_path, threshold_values)
```
