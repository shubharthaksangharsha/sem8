```python
import cv2
import matplotlib.pyplot as plt

def view_image(input_image):
    # View image using Matplotlib
    plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show();
    
def display_histogram(image, save=False):
    # Convert image to grayscale if it's a color image
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Calculate histogram
    hist = cv2.calcHist([image], [0], None, [256], [0,256])
    
    # Plot histogram
    plt.figure(figsize=(8, 6))
    plt.title("Histogram")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")
    plt.plot(hist, color='black')
    plt.xlim([0, 256])
    
    if save:
        # Save histogram as histogram.jpg
        plt.savefig('histogram.jpg')
        plt.clf()  # Clear the current figure
    else:
        plt.show()

def save_image_attributes(image):
    attributes = []
    attributes.append(f"Image shape: {image.shape}")
    attributes.append(f"Image data type: {image.dtype}")

    # Save attributes to a text file
    with open("image_attributes.txt", "w") as f:
        f.write("\n".join(attributes))

def main():
    print('Welcome to Image Processing Program')
    input_file = input("Enter input file path (e.g., input_image.jpg): ")

    # Read the input image
    input_image = cv2.imread(input_file)

    while True:
        print("\nOperations:")
        print("1. Display Image Attributes")
        print("2. Display Histogram")
        print("3. View Image")
        print("4. Save Histogram as histogram.jpg")
        print("5. Save Image Attributes as image_attributes.txt")
        print("6. Exit")
        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            # Display image attributes
            print("Image shape:", input_image.shape)
            print("Image data type:", input_image.dtype)

        elif choice == '2':
            # Display histogram
            display_histogram(input_image)
            
        elif choice == '3':
            view_image(input_image)
            
        elif choice == '4':
            # Save histogram as histogram.jpg
            display_histogram(input_image, save=True)

        elif choice == '5':
            # Save image attributes as image_attributes.txt
            save_image_attributes(input_image)

        elif choice == '6':
            print('Thank you for using...')
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 6.")

if __name__ == "__main__":
    main()
```
