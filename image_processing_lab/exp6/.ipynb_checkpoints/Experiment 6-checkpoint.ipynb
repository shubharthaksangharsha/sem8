```python
import cv2
import matplotlib.pyplot as plt

def crop_image(image, x_start, x_end, y_start, y_end):
    # Validate input coordinates
    if x_start < 0:
        x_start = 0
    if y_start < 0:
        y_start = 0
    if x_end > image.shape[1]:
        x_end = image.shape[1]
    if y_end > image.shape[0]:
        y_end = image.shape[0]
    
    # Swap coordinates if necessary
    if x_start > x_end:
        x_start, x_end = x_end, x_start
    if y_start > y_end:
        y_start, y_end = y_end, y_start
    
    cropped_image = image[y_start:y_end, x_start:x_end]
    return cropped_image


def resize_image(image, width, height):
    resized_image = cv2.resize(image, (width, height))
    return resized_image

def scale_image(image, scaling_factor):
    scaled_image = cv2.resize(image, None, fx=scaling_factor, fy=scaling_factor)
    return scaled_image

def flip_image(image, direction):
    flipped_image = cv2.flip(image, direction)
    return flipped_image

def display_images(original_image, modified_image):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(modified_image, cv2.COLOR_BGR2RGB))
    plt.title('Modified Image')
    plt.axis('off')

    plt.show()

def main():
    print('Welcome to Shubharthak Image Crop/Resize/Scale/Flip Program')
    input_file = input("Enter input file path (e.g., input_image.jpg): ")
    output_file = input("Enter output file path (e.g., output_image.jpg): ")

    # Read the input image
    input_image = cv2.imread(input_file)

#     # Display the original image
#     display_images(input_image, input_image)

    while True:
        print("\nOperations:")
        print("1. Crop")
        print("2. Resize")
        print("3. Scale")
        print("4. Flip")
        print("5. Save")
        print("6. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            x_start = int(input(f"Enter starting x-coordinate (0-{input_image.shape[1]-1}): "))
            x_end = int(input(f"Enter ending x-coordinate (0-{input_image.shape[1]-1}): "))
            y_start = int(input(f"Enter starting y-coordinate (0-{input_image.shape[0]-1}): "))
            y_end = int(input(f"Enter ending y-coordinate (0-{input_image.shape[0]-1}): "))
            output_image = crop_image(input_image, x_start, x_end, y_start, y_end)
            display_images(input_image, output_image)

        elif choice == '2':
            width = int(input("Enter width: "))
            height = int(input("Enter height: "))
            output_image = resize_image(input_image, width, height)
            display_images(input_image, output_image)

        elif choice == '3':
            scaling_factor = float(input("Enter scaling factor (e.g., 0.5): "))
            output_image = scale_image(input_image, scaling_factor)
            display_images(input_image, output_image)

        elif choice == '4':
            direction = int(input("Enter direction (0 for vertical, 1 for horizontal, -1 for both): "))
            output_image = flip_image(input_image, direction)
            display_images(input_image, output_image)

        elif choice == '5':
            cv2.imwrite(output_file, output_image)
            print(f"Modified image saved as {output_file}")
        
        elif choice == '6':
            print('Thank you for using...')
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

if __name__ == "__main__":
    main()
```
