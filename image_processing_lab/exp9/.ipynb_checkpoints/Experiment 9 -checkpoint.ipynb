### Aim: Program for morphological image operations: erosion, dilation, opening and closing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def perform_erosion(image, kernel_size=(5, 5), iterations=1):
    kernel = np.ones(kernel_size, np.uint8)
    return cv2.erode(image, kernel, iterations=iterations)

def perform_dilation(image, kernel_size=(5, 5), iterations=1):
    kernel = np.ones(kernel_size, np.uint8)
    return cv2.dilate(image, kernel, iterations=iterations)

def perform_opening(image, kernel_size=(5, 5)):
    kernel = np.ones(kernel_size, np.uint8)
    return cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

def perform_closing(image, kernel_size=(5, 5)):
    kernel = np.ones(kernel_size, np.uint8)
    return cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

def display_images(original_image, modified_image):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(modified_image, cv2.COLOR_BGR2RGB))
    plt.title('Modified Image')
    plt.axis('off')

    plt.show()

def main():
    print('Welcome to Morphological Image Operations Program')
    input_file = input("Enter input file path: ")
    
    while True:
        print("\nOperations:")
        print("1. Erosion")
        print("2. Dilation")
        print("3. Opening")
        print("4. Closing")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1': 
            image = cv2.imread(input_file)
            kernel_size = tuple(map(int, input("Enter kernel size (e.g., '5 5'): ").split()))
            iterations = int(input("Enter number of iterations: "))
            output_image = perform_erosion(image, kernel_size, iterations)
            display_images(image, output_image)
            save_option = input("Do you want to save the modified image? (y/n): ")
            if save_option.lower() == 'y':
                output_file = input("Enter output file path: ")
                cv2.imwrite(output_file, output_image)
                print(f"Erosion performed and saved as {output_file}")

        elif choice == '2':
            image = cv2.imread(input_file)
            kernel_size = tuple(map(int, input("Enter kernel size (e.g., '5 5'): ").split()))
            iterations = int(input("Enter number of iterations: "))
            output_image = perform_dilation(image, kernel_size, iterations)
            display_images(image, output_image)
            save_option = input("Do you want to save the modified image? (y/n): ")
            if save_option.lower() == 'y':
                output_file = input("Enter output file path: ")
                cv2.imwrite(output_file, output_image)
                print(f"Dilation performed and saved as {output_file}")

        elif choice == '3':
            image = cv2.imread(input_file)
            kernel_size = tuple(map(int, input("Enter kernel size (e.g., '5 5'): ").split()))
            output_image = perform_opening(image, kernel_size)
            display_images(image, output_image)
            save_option = input("Do you want to save the modified image? (y/n): ")
            if save_option.lower() == 'y':
                output_file = input("Enter output file path: ")
                cv2.imwrite(output_file, output_image)
                print(f"Opening performed and saved as {output_file}")

        elif choice == '4':
            image = cv2.imread(input_file)
            kernel_size = tuple(map(int, input("Enter kernel size (e.g., '5 5'): ").split()))
            output_image = perform_closing(image, kernel_size)
            display_images(image, output_image)
            save_option = input("Do you want to save the modified image? (y/n): ")
            if save_option.lower() == 'y':
                output_file = input("Enter output file path: ")
                cv2.imwrite(output_file, output_image)
                print(f"Closing performed and saved as {output_file}")

        elif choice == '5':
            print('Thank you for using...')
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

if __name__ == "__main__":
    main()
```
