```python
import matplotlib.pyplot as plt
import skimage.io as io
from skimage.transform import rotate

def rotate_image(image_path, angle):
    # Load the image
    image = io.imread(image_path)
    
    # Rotate the image
    rotated_image = rotate(image, angle, resize=True, mode='reflect')
    
    return image, rotated_image

image_path = "sample.jpg"
while True:
    desired_angle = int(input('Enter the desired Angle: '))  # Angle in degrees
    try:
        if desired_angle <= 0:
            print('thank you ')
            sys.exit(0)
    except Exception as e:
        pass
    
    original_image, rotated_image = rotate_image(image_path, desired_angle)
    # Display both images in subplots
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(original_image)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(rotated_image)
    plt.title(f'Rotated Image Angle: {desired_angle}')
    plt.axis('off')

    plt.show()
```

```python
import skimage.io as io
from skimage.metrics import structural_similarity as ssim
from skimage.transform import resize
import matplotlib.pyplot as plt


def calculate_ssim(image1_path, image2_path):
    # Load the images
    image1 = io.imread(image1_path, as_gray=True)
    image2 = io.imread(image2_path, as_gray=True)
    
    # Resize the images to the same dimensions
    min_height = min(image1.shape[0], image2.shape[0])
    min_width = min(image1.shape[1], image2.shape[1])
    image1_resized = resize(image1, (min_height, min_width))
    image2_resized = resize(image2, (min_height, min_width))
    
    # Calculate the Structural Similarity Index (SSI)
    ssi_value = ssim(image1_resized, image2_resized, data_range=image1_resized.max() - image1_resized.min())
    
    # Plot the images
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(image1_resized, cmap='gray')
    axes[0].set_title('Image 1')
    axes[0].axis('off')
    axes[1].imshow(image2_resized, cmap='gray')
    axes[1].set_title('Image 2')
    axes[1].axis('off')
    plt.show()
    
    return ssi_value

# Example usage:
image1_path = "sample.jpg"
image2_path = "sample2.jpg"

ssi_value = calculate_ssim(image1_path, image2_path)
print("SSI Value:", ssi_value)
```

# Draw Line

```python
from skimage import io
from skimage import draw

# Load the image
img = io.imread('sample.jpg')

# Define the coordinates for the line within the image dimensions
x, y = draw.line(0, 0, min(img.shape[0]-1, 1000), min(img.shape[1]-1, 1000))

# Draw the line on the image
img[x, y] = 0

# Display the modified image
io.imshow(img)
```

```python
from skimage import io
from skimage.draw import circle_perimeter
import numpy as np

# Load image
img = io.imread('sample.jpg')

# Calculate center coordinates of the image
x, y = img.shape[0] // 2, img.shape[1] // 2  # Center coordinates

# Define radius (adjust as needed)
radius = min(img.shape[0] // 2, img.shape[1] // 2)  # Use the smaller dimension for the radius

# Generate circle perimeter coordinates
rr, cc = circle_perimeter(x, y, radius, shape=img.shape)

# Draw circle on the image with a color that stands out
img[rr, cc] = [255, 0, 0]  # Red color for the circle, [R, G, B]
img[rr, cc] = [0, 0, 0]  # Red color for the circle, [R, G, B]

# Show image
io.imshow(img)
```

```python
from skimage import io
from skimage.draw import bezier_curve
import numpy as np

# Load image
img = io.imread('sample.jpg')

# Define Bezier curve control points
control_points = np.array([[0, 0], [500, 500], [900, 1200], [100, 100]])

# Generate Bezier curve coordinates
# x, y = bezier_curve(control_points[:, 0], control_points[:, 1], 0.5)
x, y = bezier_curve(0,0, 500, 500, 900,1200,100)


# Clip coordinates to be within image bounds
x = np.clip(x, 0, img.shape[0] - 1)
y = np.clip(y, 0, img.shape[1] - 1)

# Draw Bezier curve on the image
img[x, y] = [0, 0, 0]  # Assuming a red color for the curve, [R, G, B]

# Show image
io.imshow(img)
```
