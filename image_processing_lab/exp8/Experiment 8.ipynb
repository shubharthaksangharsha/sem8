```python
import cv2
import matplotlib.pyplot as plt

def apply_filter(image, kernel_size):
    kernel = cv2.getGaussianKernel(kernel_size, 0)
    filtered_image = cv2.filter2D(image, -1, kernel)
    return filtered_image

def display_images(original_image, filtered_image):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB))
    plt.title('Filtered Image')
    plt.axis('off')

    plt.show()

def main():
    print('Welcome to Image Filtering Program')

    # Prompt user for image file path
    input_file = input("Enter input file path (e.g., input_image.jpg): ")
    output_file = input("Enter output file path (e.g., output_image.jpg): ")

    # Read the input image
    input_image = cv2.imread(input_file)

    while True:
        print("\nFiltering Options:")
        print("1. Apply Gaussian Filter")
        print("2. Save Filtered Image")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ")

        if choice == '1':
            kernel_size = int(input("Enter the kernel size (odd number): "))
            if kernel_size % 2 == 0:
                print("Kernel size must be an odd number.")
                continue
            filtered_image = apply_filter(input_image, kernel_size)
            display_images(input_image, filtered_image)

        elif choice == '2':
            cv2.imwrite(output_file, filtered_image)
            print(f"Filtered image saved as {output_file}")

        elif choice == '3':
            print('Thank you for using Image Filtering Program.')
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 3.")

if __name__ == "__main__":
    main()
```
